```python
# Create the count plot
ax = sns.countplot(x='Size', data=df)

# Loop through the bars and add labels
for bars in ax.containers:
   
    total = len(df)  # Total number of rows in the dataset
    for bar in bars:
        height = bar.get_height()  # Get the height of each bar
        percentage = (height / total) * 100  # Calculate percentage
        ax.text(bar.get_x() + bar.get_width() / 2, height, f'{percentage:.1f}%', ha='center', va='bottom', fontsize=10, color='black')
```

### Step 4 Customer Segmentation

##### % Top 5 Sales with Lowest & Highest sales by State

```python
# Grouping sales by state
state_sales = df.groupby('ship-state')['Amount'].sum().reset_index()

# Calculating the percentage contribution of each state
total_sales = state_sales['Amount'].sum()
state_sales['Percentage'] = (state_sales['Amount'] / total_sales) * 100

# Sorting states by sales
state_sales_sorted = state_sales.sort_values(by='Amount', ascending=False)

# Top 5 states with highest sales
top_5_states = state_sales_sorted.head(5)

# Top 5 states with lowest sales
lowest_5_states = state_sales_sorted.tail(5)

# Displaying the results
print("Top 5 States with Highest Sales (in %):")
print(top_5_states[['ship-state', 'Percentage']])

print("\nTop 5 States with Lowest Sales (in %):")
print(lowest_5_states[['ship-state', 'Percentage']])
```

#### Top 5 States:

```python
# Pie chart for the top 5 states
plt.figure(figsize=(4, 4))
plt.pie(
    top_5_states['Percentage'],
    labels=top_5_states['ship-state'],
    autopct='%1.1f%%',
    colors=sns.color_palette('viridis', n_colors=5),
    startangle=140
)
plt.title('Top 5 States with Highest Sales (in %)')
plt.show()
```

#### Lowest 5 States

```python
# Pie chart for the lowest 5 states
plt.figure(figsize=(4, 4))
plt.pie(
    lowest_5_states['Percentage'],
    labels=lowest_5_states['ship-state'],
    autopct='%1.1f%%',
    colors=sns.color_palette('coolwarm', n_colors=5),
    startangle=140
)
plt.title('Top 5 States with Lowest Sales (in %)')
plt.show()
```

### Step 4 Analyzing Trends Over Time

```python
# Extracting the month from the 'Date' column
df['Month'] = df['Date'].dt.month

# Grouping sales data by month
monthly_sales = df.groupby('Month')['Amount'].sum().reset_index()

# Sorting months for proper visualization
monthly_sales = monthly_sales.sort_values(by='Month')

# Displaying the grouped data
monthly_sales
```

```python
import calendar
# Check the format of the Month column
if monthly_sales['Month'].dtype == 'int64' or monthly_sales['Month'].dtype == 'float64':
    # Convert numeric months to full month names
    monthly_sales['Month'] = monthly_sales['Month'].apply(lambda x: calendar.month_name[int(x)])
elif monthly_sales['Month'].dtype == 'object':
    print("Months are already in string format.")

# Plotting the monthly sales trend
plt.figure(figsize=(6, 3))
ax = sns.lineplot(x='Month', y='Amount', data=monthly_sales, marker='o')

# Adding labels and title
ax.set_xlabel('Month')
ax.set_ylabel('Total Sales Amount')
ax.set_title('Monthly Sales Trend')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)
plt.show()
```

## Step 6. Correlation Analysis

```python
# Selecting relevant numerical columns for correlation analysis
corr_columns = ['Amount', 'Qty']

# Calculate the correlation matrix
corr_matrix = df[corr_columns].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(4, 2))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Sales Data')
plt.show()
```

##### The weak correlation between Qty and Amount implies that the sales value isn't strictly dependent on the quantity sold in your dataset. The correlation could be higher if other factors, like pricing or product type, were considered.

### Exploration of more Correlations

```python
# First, let's ensure we are selecting only numeric columns for correlation calculation
df_encoded = df.copy()

# Convert categorical columns to numeric codes
df_encoded['Category'] = df_encoded['Category'].astype('category').cat.codes
df_encoded['Size'] = df_encoded['Size'].astype('category').cat.codes
df_encoded['Status'] = df_encoded['Status'].astype('category').cat.codes
df_encoded['ship-state'] = df_encoded['ship-state'].astype('category').cat.codes

# Select only numeric columns for correlation
df_numeric = df_encoded.select_dtypes(include=['number'])

# Calculate correlations
correlation_matrix = df_numeric.corr()

# Display the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix between Numeric Variables")
plt.show()
```
